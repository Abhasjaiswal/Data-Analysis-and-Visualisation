```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
import seaborn as sns 
```

```python
df=pd.read_csv('./datasets/iris.csv')
```

```python
df
```

```python
df.head(5)
```

```python
df.dtypes
```

```python
df.drop("Species",axis=1,inplace=True)
df.head(30)
```

```python
df.drop(2,axis=0,inplace=True)
```

```python
df.drop([3, 5, 10], axis=0,inplace=True)
df.head(10)
```

```python
df.drop(index=[4, 7, 11], columns=['SepalWidthCm', 'PetalWidthCm'],inplace=True)
df.head(10)
```

```python
df=df.rename(columns={'SepalLengthCm':'SLcm','PetalLengthCm':'PLcm'})
df.head(10)
```

```python
df.duplicated().head(10)
```

```python
duplicate_rows = df[df.duplicated()]
print("number of duplicate rows : ", duplicate_rows.shape)
```

```python
df.count()
```

```python
df.isnull().sum()
```

```python
df=df.dropna()
df.count()
```

```python
df
```

```python
sns.boxplot(data=df)
plt.show()
```

```python
sns.boxplot(x=df["SLcm"])
plt.show()
```

```python
plt.hist(df['SLcm'], bins=10, edgecolor='black')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Frequency')
plt.title('Histogram of Sepal Length')
plt.show()
```

```python
for column in df.select_dtypes(include=['number']):
    plt.hist(df[column])
    plt.xlabel(column)
    plt.ylabel('Frequency')
    plt.title(f'Histogram of {column}')
    plt.show()
```

```python
correlation_matrix = df.corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.show()
```

```python
sns.scatterplot(x='SLcm', y='PLcm', data=df)
plt.show()
```
